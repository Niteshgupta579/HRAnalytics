```python
x=df['Salary'].sort_values(ascending=False).head(10)
y= df['Salary'].sort_values(ascending= False).tail(10)

c = np.arange(len(x)) # Define c as numerical positions for the bars

plt.bar(c,x,color='g',label='Highest salaries')
plt.bar(c,y,color='r',label='Lowest salaries')

plt.title('top 10 higest vs lowest salaries',fontsize=10)
plt.xticks(c) # Set x-ticks to correspond to the bar positions
plt.ylabel('salaries')
plt.legend()
plt.show()
```

```python
#lowest salary are in range
```

```python
import pandas as pd
df=pd.read_csv('HRDataset_v14 .csv')
df['RecruitmentSource'].unique()
```

```python
df['RecruitmentSource'].value_counts()
```

```python
import matplotlib.pyplot as plt

recruitment_counts = df['RecruitmentSource'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(recruitment_counts.index, recruitment_counts.values, color='skyblue')
plt.title('Source of Recruitment', fontsize=14)
plt.xlabel('Recruitment Source', fontsize=12)
plt.ylabel('Number of Candidate Hires', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
#insights
#indeed is the most common
```
